#### getting data

```python
import os
import xml.etree.ElementTree as ET
import cv2
import matplotlib.pyplot as plt
import random
import warnings
warnings.filterwarnings("ignore")
```

```python
directory = os.getcwd()
images_dir = os.path.join(directory,'Images')
annotation_dir = os.path.join(directory,'Annotations')
folders_assigned=os.listdir(images_dir)
```

```python
folders_assigned
```

```python
target_size = (100, 100)
#cropped data
cropped='Cropped'
os.makedirs(cropped, exist_ok=True)
```

```python
#check if file is xml
def is_xml(file_path):
    try:
        with open(file_path, 'r') as file:
            content = file.read()
            return '<annotation>' in content and '</annotation>' in content
    except Exception as e:
        return False
```

```python
for folder in folders_assigned:
    images = os.path.join(images_dir, folder)
    annotations = os.path.join(annotation_dir, folder)
    for file in os.listdir(annotations):
        path = os.path.join(annotations, file)
        #print(file)
        if is_xml(path):
            tree = ET.parse(path)
            root = tree.getroot()
            image_filename = root.find('filename').text
            image_path = os.path.join(directory,images, image_filename)
            #print(image_path)
            image = cv2.imread(image_path+".jpg")
            for obj in root.findall('.//object'):
                bbox = obj.find('bndbox')
                xmin = int(bbox.find('xmin').text)
                ymin = int(bbox.find('ymin').text)
                xmax = int(bbox.find('xmax').text)
                ymax = int(bbox.find('ymax').text)
                roi = image[ymin:ymax, xmin:xmax]
                crop = cv2.resize(roi, target_size)
                req=os.path.join(cropped,folder)
                os.makedirs(req, exist_ok=True)
                output_path = os.path.join(cropped,folder, f'{image_filename}.jpg')
                #print(output_path)
                cv2.imwrite(output_path, crop)
```

### histogram equalization

```python
img=[]
for folder in os.listdir(cropped):
    path = os.path.join(cropped, folder)
    if os.path.isdir(path):
        images = os.listdir(path) 
        crop_images = [image for image in images if image.lower().endswith(('.jpg'))][:2]
        for image in crop_images:
            src_path = os.path.join(path, image)
            img.append(src_path)
```

```python
img
```

```python
#plot grey images
def plot_grey(name,grey_image):
    plt.figure(figsize=(8, 4))
    plt.subplot(1, 2, 1)
    plt.imshow(grey_image, cmap='gray')
    plt.title('Greyscale Image')
    plt.subplot(1, 2, 2)
    plt.hist(grey_image.ravel(), bins=256, range=[0, 256], density=True, color='red', alpha=0.7)
    plt.title('Pixel Intensity Histogram')
    plt.xlabel('Pixel Intensity')
    plt.ylabel('Frequency')           
    plt.suptitle(f'{name}')
    plt.show() 
    
```

```python
def plot_equa(name,equ):
    plt.figure(figsize=(8, 4))
    plt.subplot(2, 2, 1)
    plt.imshow(equ, cmap='gray')
    plt.title('Equalized Greyscale Image')
    plt.subplot(2, 2, 2)
    plt.hist(equ.ravel(), bins=256, range=[0, 256], density=True, color='b', alpha=0.7)
    plt.title('Equalized Pixel Intensity Histogram')
    plt.xlabel('Pixel Intensity')
    plt.ylabel('Frequency')
    plt.suptitle(f'{name}')
    plt.show()

```

```python
for image in img:
    #print(image)
    img = cv2.imread(image)
    grey_img = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
    # Perform histogram equalization
    equalized_img = cv2.equalizeHist(grey_img)
           
    plot_grey(image,grey_img)  
    plot_equa(image,equalized_img)
    plt.subplot(221), plt.imshow(grey_img, cmap='gray')
    plt.subplot(222), plt.imshow(equalized_img, cmap='gray')
    plt.show()
    

```

### histogram comparision

Reference : https://stackoverflow.com/questions/67323056/histogram-comparison-between-two-images

```python
imag=['Cropped\\n02091134-whippet\\n02091134_10107.jpg',
 'Cropped\\n02091134-whippet\\n02091134_10219.jpg',
 'Cropped\\n02099601-golden_retriever\\n02099601_10.jpg']
```

```python
hist_Sizes = [256]
ranges = [0, 256]
hist=[]
for img in imag:
    imge=cv2.imread(img)
    gray=cv2.cvtColor(imge,cv2.COLOR_BGR2GRAY)
    hist1 = cv2.calcHist([gray], [0], None, hist_Sizes, ranges)
    hist.append(hist1)
    
```

```python
def comparision(hist1,hist2):
    print("Manhattan Distance: {}".format(cv2.norm(hist1, hist2, normType=cv2.NORM_L1)))
    print("Euclidean Distance: {}".format(cv2.norm(hist1, hist2, normType=cv2.NORM_L2)))
    print("Bhattacharyya Distance: {}".format(cv2.compareHist(hist1, hist2, method = cv2.HISTCMP_BHATTACHARYYA)))
    print("Histogram Intersection: {}".format(cv2.compareHist(hist1, hist2, method = cv2.HISTCMP_INTERSECT)))
```

```python
print("comparision between images of same class\n")
comparision(hist[0],hist[1])
```

```python
print("comparision between images of different classes\n")

comparision(hist[1],hist[2])
```

### PCA

Reference : https://scikit-learn.org/stable/modules/generated/sklearn.decomposition.PCA.html

```python
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler
hist_Sizes = [256]
ranges = [0, 256]
n= []
hist = []
for i,name in enumerate(os.listdir(images_dir)[:2]):
    path = os.path.join(images_dir, name)
    if os.path.isdir(path):
        files = os.listdir(path)
        images = [file for file in files if file.lower().endswith(('.jpg'))]
        for image in images:
            src = os.path.join(path, image)
            img = cv2.imread(src)
            grey = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
            ghist = cv2.calcHist([grey], [0], None, hist_Sizes, ranges)
            scale=StandardScaler()
            gghist=scale.fit_transform(ghist)            
            hist.append(gghist.flatten())
            n.append(i)
```

```python
pca=PCA(2)
points = pca.fit_transform(hist)
plt.figure(figsize=(10,10))
plt.scatter(points[:,0],points[:,1],c=n,cmap='plasma')
plt.xlabel('pc1')
plt.ylabel('pc2')
```
